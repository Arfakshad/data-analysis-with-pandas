```python
import numpy as np 
import pandas as pd 
```

#### 1. Creating Data with Missing Values

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

#### 2. Finding Missing Data

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

#### 3. Removing Missing Data

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=1)
```

#### 4. Filling Missing Data

```python
df.fillna(0)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
