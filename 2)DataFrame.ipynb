### 🔹 What is a DataFrame?

##### A DataFrame in Pandas is a two-dimensional table-like data structure used to store and organize data.

##### 👉 Think of it like:

##### An Excel sheet 📊
##### Or a database table

### 1. Import Libraries

```python
import numpy as np 
import pandas as pd 
```

#### 2. Creating DataFrame from Dictionary

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

#### 3. Creating DataFrame from List

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
```

```python
df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

```python
df2
```

#### 4. Selecting Columns

```python
df2["City"]
```

```python
df2[["Name","City"]]
```

#### 5. Creating a New Column

```python
df2["Roll"] = ["Doctor","Eng.","Doctor","Eng."]
```

```python
df2["Designation"] = ["Doctor","Eng.","Doctor","Eng."]
```

```python
df2
```

#### 6. Removing Rows / Columns

```python
df2.drop(0, axis=0)
```

```python
df2.drop("Roll", axis=1)
```

```python
df2.drop("Designation", axis=1)
```

```python
df2.drop(["Designation","Roll"], axis=1, inplace=True)
```

```python
df2
```

#### 7. Selecting Rows

```python
df2.loc[[0,1]]
```

```python
df.iloc[3]
```

#### 8. Selecting Rows + Columns

```python
df.loc[[0,1]][["City","Salary"]]
```

#### 9. Conditional Selection (Filtering)

```python
df2
```

```python
#I only want to see those people whose age is above 30
```

```python
df2[df2["Age"] > 30]
```

```python
#I only want poeple whose age is above 30 and their city must be paris
```

```python
df2[(df2["Age"] > 30) & (df2["City"] == 'Paris')]
```
