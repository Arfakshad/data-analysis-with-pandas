### What is a Pandas Series?

##### A Pandas Series is a one-dimensional labeled array:

##### It stores data (like a list)
##### Each value has a label (index)

#### 1.Libraries


```python
import numpy as np 
import pandas as pd 
```

#### 2. Creating Data

```python
labels = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
```

#### 3. Create Series from List

```python
pd.Series(my_list)
```

#### 4. Series with Custom Index

```python
pd.Series(my_list,index=labels)
```

#### 5. Series from NumPy Array

```python
pd.Series(arr)
```

#### 6. Series from Dictionary

```python
pd.Series(d)
```
